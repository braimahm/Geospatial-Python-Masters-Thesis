# Detector Offset Function

```python
"""
This function corrects the detector offset
 """
def detector_offset_correction(reflectance):

   #first drop the wavelength
  subset = reflectance.drop('Wavelength', axis = 1)

  #store 350nm to 1000nm as top_rows (index 0 to 650)
  top_rows = subset.iloc[0:651]

  #calculate the correction factor at 1000 nm/1001 nm
  c1 = (subset.iloc[650])/(subset.iloc[651])

  #calculate the correction factor at 1800 nm/1801 nm
  c2 = (subset.iloc[1450])/(subset.iloc[1451])

  #multiply the rows beyond 1000nm by the correction factor (c1)
  c1_1000 = c1* (subset.iloc[651:])

  #Store the corrected rows from 1001nm to 1800nm as corrected_1
  corrected_1 = pd.concat([ top_rows, c1_1000], ignore_index=False)

  #  #multiply values beyond 1800nm by the correction factor (c2)
  corrected_2 = c2* (corrected_1.iloc[1451:])

  # #Store the second corrected rows as corrected_2


  # #Join the two set of corrected rows
  corrected_final = pd.concat([corrected_1.iloc[0:1451], corrected_2], ignore_index=False)

 #now insert the Wavelength and return the final results
  corrected_final.insert(loc = 0, column = 'Wavelength', value = reflectance['Wavelength'])

  return corrected_final
```
